**Developing a Scheduler for Multiprocessing** <br>
Objective of this lab is to implement an automatic scheduler for assigning jobs to the CPU-cores based on the size of jobs.

```python
# various helper functions to compute according to the given time complexity.
```

```python
def make_array(n):
  import random 
  arr=[]
  for i in range(n):
    arr.append(random.randint(1,11))
  return arr
```

```python
def compute_n():
  for i in range(n):
    print(i)
```

```python
def compute_n2():
  arr = make(n)
  for i in range(len(arr)): 

      min_idx = i 
      for j in range(i+1, len(arr)): 
          if arr[min_idx] > arr[j]: 
              min_idx = j        
      arr[i], arr[min_idx] = arr[min_idx], arr[i] 
```

```python
def compute_n3():
  for i in range(n):
    for j in range(n):
      for k in range(n):
        print(i,j,k, end = "")
```

```python
def compute_logn():
  import math
  for i in range(int(math.log(n))):
    print(i)
```

```python
def compute_nlogn():
  for i in range(n):  
    for j in range(int(math.log(n))):
      print(i,j, end = " ")
```

```python
# array size
import math
n = 25
functions = [compute_n, compute_n2, compute_n3, compute_logn, compute_nlogn]

time_dict = {}
import time
for func in functions:
  start_time = time.time();
  func()
  time_dict[func.__name__] = (time.time() - start_time)
```

```python
timer = []

from multiprocessing import Process
start_time = time.time()
for i in range(0,len(functions),2):
  processes = [Process(target=functions[i])]
  try :
    processes.append(Process(target=functions[i+1]))
  except:
    print("Last Process")
  print(processes)
  for process in processes:
        process.start()

  for process in processes:
        process.join()
timer.append(time.time() - start_time)
```

```python
# knapsack

functions.reverse()

from multiprocessing import Process
start_time = time.time()
for i in range(0,len(functions),2):
  processes = [Process(target=functions[i])]
  try :
    processes.append(Process(target=functions[i+1]))
  except:
    print("Last Process")
  print(processes)
  for process in processes:
        process.start()

  for process in processes:
        process.join()
timer.append(time.time() - start_time)
```
